```python
subway.head()
```

```python
subway = subway.rename(columns = {
    '서강지구대_승차' : 'seogang_on',
    '공덕지구대_승차' : 'gongdeok_on',
    '월드컵지구대_승차' : 'worldcup_on',
    '홍익지구대_승차' : 'hongik_on',
    '용강지구대_승차' : 'yonggang_on'
})
subway.columns
```

```python
subway = subway.rename(columns = {
    '서강지구대_하차' : 'seogang_off',
    '공덕지구대_하차' : 'gongdeok_off',
    '월드컵지구대_하차' : 'worldcup_off',
    '홍익지구대_하차' : 'hongik_off',
    '용강지구대_하차' : 'yonggang_off'
})
subway.columns
```

```python
predict112 = pd.merge(predict112, subway, on='date', how='inner') # 건수, 승하차, 요일 결합
predict112.head()
```
